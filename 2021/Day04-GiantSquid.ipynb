```python
import numpy as np

# Read numbers from text file (stripping newline chars and converting to int)
data = open("inputs/04-input.txt").read().splitlines()

numbers = np.array([int(x) for x in data[0].split(',')])

cards = [np.array(' '.join(data[(6*i)+2:(6*i)+7]).split()).reshape(5,5) for i in range(0,len(data)//6)]

cards_list = [list(map(int,' '.join(data[(6*i)+2:(6*i)+7]).split())) for i in range(0,len(data)//6)]
cards_list

def bingo(numbers:list, card:list):
    setn = set(numbers)
    setc = set(card) 
    if len(setn & setc)>=5:
        matches = [n in numbers for n in card]
        sum_non_matches = sum(card[i] for i,val in enumerate(matches) if not val)
        # Any matching row?
        
        if any([all(matches[5*i:5*i+5]) for i in range(0,5)]):
            return (True, sum_non_matches)
        if any([all(matches[i::5]) for i in range(0,5)]):
            return (True, sum_non_matches)
        return False
    else:
        return False

for i in range(0,len(numbers)):
    for j in range(0,len(cards_list)):
        is_bingo = bingo(numbers[:i], cards_list[j])
        if is_bingo:
            print(i,j)
            print('Card: \n',np.array(cards_list[j]).reshape(5,5))
            print('Numbers: \n', numbers[:i])
            answer = is_bingo[1] * numbers[i-1]
            print(f'Sum non-matches ({is_bingo[1]}) x last number called ({numbers[i-1]}) = {answer}')

            break
    if is_bingo:
        break
```

## Part 2

```python
for i in range(len(numbers),0,-1):
    for j in range(0,len(cards_list)):
        is_bingo = bingo(numbers[:i], cards_list[j])
        if not is_bingo:
            print('Card: \n',np.array(cards_list[j]).reshape(5,5))
            print('Numbers: \n', numbers[:i])
            print(f'Sum non-matches ({bingo(numbers[:i+1], cards_list[j])[1]}) x last number called ({numbers[i]}) = {bingo(numbers[:i+1], cards_list[j])[1] * numbers[i]}')
            break
    if not is_bingo:
        break
```
