```python
from collections import Counter

input = open("inputs/23-input.txt").read().splitlines()

# input = """....#..
# ..###.#
# #...#.#
# .#...##
# #.###..
# ##.#.##
# .#..#..""".splitlines()

elves = {complex(x,y) for y,r in enumerate(reversed(input)) for x, c in enumerate(r) if c =="#"}

# Unnecessary function to show map and help me find my stupid errors
def show(elves):
    x,y = zip(*[(int(e.real),int(e.imag)) for e in elves])
    rows = [ ["." for _ in range(max(x)-min(x)+1,0,-1)] for _ in range(max(y)-min(y)+1)]

    for e in elves:
        rows[int(e.imag-min(y))][int(e.real-min(x))] = "#" 

    print("\n".join("".join(r) for r in reversed(rows)))

# N, S, W, E
dirs = (1j, -1j, -1, 1)
# NE, SE, SW, NW
diags = (1+1j, 1-1j, -1-1j, -1+1j)

adj = dirs + diags

# Are all 3 adjacent positions in direction d clear?
def move_clear(e,d,elves):
    return not(any(e+d + d*x in elves for x in (0,1j,-1j)))

for r in range(1,1000):
    # If adjacent positions clear, choose move
    def move(e):
        if any(e + a in elves for a in adj):
            return next((e+d for d in dirs if move_clear(e,d,elves)), None)

    # Proposed moves and counts of each
    proposed = list(map(move, elves))
    counts = Counter(proposed)

    # start/end positions (for elves with unique proposed position)
    moves = [(start,end) for start,end in zip(elves, proposed) if counts[end] == 1]

    if not moves:
        print(r)
        break

    # update elves (remove start positions, add end positions)
    start,end = map(set,zip(*moves))
    elves = elves - start | end

    # Shuffle directions
    dirs = dirs[1:] + dirs[:1]

    # Check free space after 10 rounds
    if r==10:
        show(elves)
        x,y = zip(*[(e.real,e.imag) for e in elves])
        empty = (max(y) - min(y) + 1) * (max(x) - min(x) + 1) - len(elves)
        print(f"{empty} empty ground tiles")
```
